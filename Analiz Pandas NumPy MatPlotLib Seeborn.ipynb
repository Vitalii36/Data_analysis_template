```python
job_df = df[['MORTDUE', 'VALUE', 'YOJ', 'DELINQ', 'CLAGE', 'NINQ'] + ['DEROG']]
job_df.groupby('DEROG').sum().plot()
```

#### Plotli

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
data = []
for genre in df.JOB.unique():
    data.append(
        go.Box(y=df[df.JOB==genre].VALUE, name=genre)
    )

# визуализируем данные
iplot(data, show_link = False)
```

```python
features = list(set(df.columns) - set(['MORTDUE', 'VALUE', 'YOJ', 'DELINQ', 'CLAGE', 'NINQ']))

df[features].hist(figsize=(20,12));
```
